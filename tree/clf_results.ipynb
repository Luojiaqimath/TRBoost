```python
import numpy as np
import os
import pandas as pd
```

# clf standard

```python
folder_path = "./clf_standard_result"
data_list = ['adult', 'german','electricity',
             'sonar', 'credit', 'spam']
```

```python
# trb
stad_df1 = pd.DataFrame(columns=['trb_auc_mean', 'trb_auc_std', 'trb_f1_mean', 'trb_f1_std',
                                 'trb_est_mean', 'trb_ets_std'],
                        index=data_list)
for i, dataset in enumerate(data_list):
    data = dict(np.load(os.path.join(folder_path, dataset, 'metric_lists.npz')))
    stad_df1.iloc[i:(i+1), :] = data['trb_auc'].mean(),data['trb_auc'].std(),\
        data['trb_f1'].mean(),data['trb_f1'].std(), data['trb_nestimator'].mean(),data['trb_nestimator'].std()
```

```python
# xgb
stad_df2 = pd.DataFrame(columns=['xgb_auc_mean', 'xgb_auc_std', 'xgb_f1_mean', 'xgb_f1_std',
                                 'xgb_est_mean', 'xgb_ets_std'],
                        index=data_list)
for i, dataset in enumerate(data_list):
    data = dict(np.load(os.path.join(folder_path, dataset, 'metric_lists.npz')))
    stad_df2.iloc[i:(i+1), :] = data['xgb_auc'].mean(),data['xgb_auc'].std(),\
        data['xgb_f1'].mean(),data['xgb_f1'].std(),data['xgb_nestimator'].mean(),data['xgb_nestimator'].std()
```
